<h1>Explore the Î¸-rule for exponential growth</h1>



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def exact(x):
    return np.exp(-x)

def func(x,theta,dt):
    u = np.zeros(len(x))
    u[0] = 1
    for i in range(len(x)-1):
        u[i+1] = u[i]*( (1-(1-theta)*dt) / (1+theta*dt) )
    return u
```

```python
thetas = [0,0.5,1]
dts = [1,0.5,0.1,0.01]
```

```python
for theta in thetas:
    for dt in dts:
        x = np.arange(0,5,dt)
        u = func(x,theta,dt)
        plt.plot(x,u,label=r'$\theta$ = {}, $\Delta t$ = {}'.format(theta,dt))
    plt.plot(x,exact(x),label='exact')
    plt.legend()
    plt.show()
```

For all small $Î t$ the models perform well, and align with the exact solution. Its more interesting to fint the characteristics og the models by seeing how it acts when given larger $Î t$'s.

Forward Euler ($Î¸ = 0$):
-   For large $Î t$ the model diverges from the exact solution, and the given solution always lies above the exact solution. 

Backward Euler ($Î¸ = 1$):
-   For large $Î t$ the model  perfomes better than the Forward Euler model, but still diverges from the exact solution. The given solution always lies below the exact solution.

Crank-Nicolson ($Î¸ = 0.5$):
-   For large $Î t$ the model perfomes the best of the three, and aligns pretty well with the exact solution. The given solution lies above and below the exact solution, but the difference is smaller than the other two models.

```python
def A_exact(p):
    return np.exp(-p)

def A(p, theta):
    return (1-(1-theta)*p)/(1+theta*p)
```

```python
def amplification_factor():
    p = np.linspace(0,3,100)
    exact =  A_exact(p)
    plt.plot(p,exact,label='exact')
    for theta in thetas:
        curve = A(p,theta)
        plt.plot(p,curve,label=r'$\theta$ = {}'.format(theta))
    
    plt.plot(p,np.zeros(len(p)),'k--')#A=0 line
    plt.legend()
    plt.show()

amplification_factor()
```
